```python
import tkinter as tk
from tkinter import *
from tkinter import ttk
from tkinter import filedialog
import pandas as pd
import re
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import ImageTk,Image
import pickle
import tensorflow
from tensorflow.keras.preprocessing.image import img_to_array
import keras.models
from keras.models import load_model
import warnings
warnings.filterwarnings("ignore")
```

```python
model_df = pd.read_csv('modelmetrics.csv')
```

```python
def preprocess_img():
    Original = Image
    original = Image.open(image_data)
    original = original.resize((224, 224), Image.ANTIALIAS)
    numpy_image = img_to_array(original)
    image_batch = np.expand_dims(numpy_image, axis=0)
    return image_batch
```

```python
def load_img():
    global img, image_data
    for img_display in frame.winfo_children():
        img_display.destroy()

    image_data = filedialog.askopenfilename(initialdir="/", title="Choose an image",
                                       filetypes=(("all files", "*.*"), ("jpg files", "*.jpg")))
    basewidth = 150 # Processing image for dysplaying
    img = Image.open(image_data)
    wpercent = (basewidth / float(img.size[0]))
    hsize = int((float(img.size[1]) * float(wpercent)))
    img = img.resize((basewidth, hsize), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(img)
    file_name = image_data.split('/')
    panel = tk.Label(frame, text= str(file_name[len(file_name)-1]).upper()).pack()
    panel_image = tk.Label(frame, image=img).pack()
```

```python
def classifyknn():
    original = Image.open(image_data)
    original = original.resize((224, 224), Image.ANTIALIAS)
    numpy_image = img_to_array(original)
    image_batch = np.expand_dims(numpy_image, axis=0)
    image_batch = image_batch.reshape(1,(224*224*3))
    pick_model = open('KNeighbours.pkl', 'rb')
    model = pickle.load(pick_model)
    pick_model.close()
    print(image_batch.shape)
    predictions = model.predict(image_batch)
    lbltext = "Model Name: "+ model_df.loc[(model_df['Model']=="KNN")]['Model'][0] + "\n\n" 
    lbltext = lbltext + "Train Accuracy: "+ str(model_df.loc[(model_df['Model']=="KNN")]['Train Accuracy'][0]) + "\n\n" 
    lbltext = lbltext + "Test Accuracy: "+ str(model_df.loc[(model_df['Model']=="KNN")]['Test Accuracy'][0]) + "\n\n" 
    lbltext = lbltext + " The Prediction Label is: " + str(np.argmax(predictions))
    label1 = ttk.Label(text = lbltext)
    label1.place(x=850, y=350)
    Image_batch = []
```

```python
def classifyRF():
    original = Image.open(image_data)
    original = original.resize((224, 224), Image.ANTIALIAS)
    numpy_image = img_to_array(original)
    image_batch = np.expand_dims(numpy_image, axis=0)
    image_batch = image_batch.reshape(1,(224*224*3))
    pick_model = open('RandomForest.pkl', 'rb')
    model = pickle.load(pick_model)
    pick_model.close()
    predictions = model.predict(image_batch)
    lbltext = "Model Name: "+ model_df.loc[(model_df['Model']=="RandomForest")]['Model'][1] + "\n\n" 
    lbltext = lbltext + "Train Accuracy: "+ str(model_df.loc[(model_df['Model']=="RandomForest")]['Train Accuracy'][1]) + "\n\n" 
    lbltext = lbltext + "Test Accuracy: "+ str(model_df.loc[(model_df['Model']=="RandomForest")]['Test Accuracy'][1]) + "\n\n" 
    lbltext = lbltext + " The Prediction Label is: " + str(np.argmax(predictions))
    label1 = ttk.Label(text = lbltext)
    label1.place(x=650, y=350)
    Image_batch = []
```

```python
def classifyNN():
    original = Image.open(image_data)
    original = original.resize((224, 224), Image.ANTIALIAS)
    numpy_image = img_to_array(original)
    image_batch = np.expand_dims(numpy_image, axis=0)
    image_batch = image_batch.reshape(1, (224*224*3))
    model = keras.models.load_model('flowers_NN.h5')
    predictions = model.predict(image_batch)
    lbltext = "Model Name: "+ model_df.loc[(model_df['Model']=="Neural Networks")]['Model'][2] + "\n\n" 
    lbltext = lbltext + "Train Accuracy: "+ str(model_df.loc[(model_df['Model']=="Neural Networks")]['Train Accuracy'][2]) + "\n\n" 
    lbltext = lbltext + "Test Accuracy: "+ str(model_df.loc[(model_df['Model']=="Neural Networks")]['Test Accuracy'][2]) + "\n\n" 
    lbltext = lbltext + " The Prediction Label is: " + str(np.argmax(predictions))
    label1 = ttk.Label(text = lbltext)
    label1.place(x=450, y=350)
    Image_batch = []
```

```python
def classifyCNN():
    original = Image.open(image_data)
    original = original.resize((224, 224), Image.ANTIALIAS)
    numpy_image = img_to_array(original)
    image_batch = np.expand_dims(numpy_image, axis=0)
    model = keras.models.load_model('flowers_CNN.h5')
    predictions = model.predict(image_batch)
    model.summary()
    lbltext = "Model Name: "+ model_df.loc[(model_df['Model']=="CNN")]['Model'][3] + "\n\n" 
    lbltext = lbltext + "Train Accuracy: "+ str(model_df.loc[(model_df['Model']=="CNN")]['Train Accuracy'][3]) + "\n\n" 
    lbltext = lbltext + "Test Accuracy: "+ str(model_df.loc[(model_df['Model']=="CNN")]['Test Accuracy'][3]) + "\n\n" 
    lbltext = lbltext + " The Prediction Label is: " + str(np.argmax(predictions))
    label1 = ttk.Label(text = lbltext)
    label1.place(x=280, y=350)
    Image_batch = []
```

```python
def classifyCNN_Resnet():
    original = Image.open(image_data)
    original = original.resize((224, 224), Image.ANTIALIAS)
    numpy_image = img_to_array(original)
    image_batch = np.expand_dims(numpy_image, axis=0)
    model = keras.models.load_model('17flowers_resnet.h5')
    model.summary()
    predictions = model.predict(image_batch)
    lbltext = "Model Name: "+ model_df.loc[(model_df['Model']=="CNN-Resnet")]['Model'][4] + "\n" 
    lbltext = lbltext + "Train Accuracy: "+ str(model_df.loc[(model_df['Model']=="CNN-Resnet")]['Train Accuracy'][4]) + "\n\n" 
    lbltext = lbltext + "Test Accuracy: "+ str(model_df.loc[(model_df['Model']=="CNN-Resnet")]['Test Accuracy'][4]) + "\n\n" 
    lbltext = lbltext + " The Prediction Label is: " + str(np.argmax(predictions))
    label1 = ttk.Label(text = lbltext)
    label1.place(x=120, y=350)
    Image_batch = []
```

```python
root = tk.Tk()
root.title('Portable Image Classifier')
root.resizable(False, False)
tit = tk.Label(root, text="Portable Image Classifier", padx=25, pady=6, font=("", 12)).pack()
canvas = tk.Canvas(root, height=500, width=500, bg='grey')
canvas.pack()
frame = tk.Frame(root, bg='white')
frame.place(relwidth=0.8, relheight=0.8, relx=0.1, rely=0.1)
chose_image = tk.Button(root, text='Choose Image',
                        padx=35, pady=10,
                        fg="white", bg="grey", command=load_img)
chose_image.pack(side=tk.LEFT)
class_image = tk.Button(root, text='Classify Image-KNN',
                        padx=35, pady=10,
                        fg="white", bg="grey", command=classifyknn)
class_image.pack(side=tk.RIGHT)
class_image = tk.Button(root, text='Classify Image-RF',
                        padx=35, pady=10,
                        fg="white", bg="grey", command=classifyRF)
class_image.pack(side=tk.RIGHT)
class_image = tk.Button(root, text='Classify Image-NN',
                        padx=35, pady=10,
                        fg="white", bg="grey", command=classifyNN)
class_image.pack(side=tk.RIGHT)
class_image = tk.Button(root, text='Classify Image-CNN',
                        padx=35, pady=10,
                        fg="white", bg="grey", command=classifyCNN)
class_image.pack(side=tk.RIGHT)
class_image = tk.Button(root, text='Classify Image-CNN-Resnet',
                        padx=35, pady=10,
                        fg="white", bg="grey", command=classifyCNN_Resnet)
class_image.pack(side=tk.RIGHT)
root.mainloop()
```

### Strategy to maintain this in production

### With respect to Transfer learning, the process is faster and accurate, while using less number of training data(Trainable params: 1,089,553) , while this gets 3 times higher while using conventional CNN with image augmentation and further up for NN. Even if the data grows the model will be able to handle it with the current trainable parameters, which in-turn reduces the training time as well
