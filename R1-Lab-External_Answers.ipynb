## About the dataset

This is perhaps the best known database to be found in the pattern recognition literature. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant.  One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

Attribute Information:
   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm
   4. petal width in cm
   5. class: 
      -- Iris Setosa
      -- Iris Versicolour
      -- Iris Virginica

### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
RangeSepallength = iris[['Sepal Length (in cm)']].max() - iris[['Sepal Length (in cm)']].min()
RangeSepalWidth = iris[['Sepal Width in (cm)']].max() - iris[['Sepal Width in (cm)']].min()
RangePetallength = iris[['Petal length (in cm)']].max() - iris[['Petal length (in cm)']].min()
RangePetalWidth = iris[['Petal width (in cm)']].max() - iris[['Petal width (in cm)']].min()
RangeSepalExtra = iris[['SepalExtra']].max() - iris[['SepalExtra']].min()
print(RangeSepallength)
print(RangeSepalWidth)
print(RangePetallength)
print(RangePetalWidth)
print(RangeSepalExtra)
iris.describe(include="all")  ### will give summary statistics for all variables - 5 Point Summary
```
